#### 1. Introduction: The diet problem. 

The diet problem arises in the 
context of World War II. The navy wanted
to minimize the cost of feeding without prejudice 
of offer a balanced diet to soldiers. 


<img src="img/Foodisammunition.jpg" alt="CAChemE" style="width: 400px;"/>
<center><sup><a href="http://www.neos-guide.org/content/diet-problem" target="_blank">Source: Neos Server</a></sup></center>
 

One of the first researchers to study the problem was 
[George Stigler](https://en.wikipedia.org/wiki/George_Stigler),
who made a prediction for the value of an optimal menu ($39.93)
in 1939 using heuristic methods. 

In 1947, the team of 
[Jack Ladermann](https://www.researchgate.net/researcher/2017944516_Jack_Laderman) 
used the SIMPLEX method to solve the Stigler's model, wich
was formed from 9 equations with 77 unknows. 
The work was realized by 9 employees during 120 days and 
giving a result of $39.69, it seems that Stigler was right!.

#### 5. Types of optimization problems and modelling forms. 

<img src="img/op_types1.png" alt="CAChemE" style="width: 2500px;"/>

An optimization model can be defined mathematically as follows:

$$
\begin{array}{lll}
\min & f(x) = \sum_{i=1}^F c_i x_i & \\
s.t. & \sum_{i=1}^F \sum_{j=1}^N a_{ij} x_i \leq N_{max} & \forall i = 1 \ldots F\ & \forall j = \ldots N \\
     & \sum_{i=1}^F \sum_{j=1}^N a_{ij} x_i \geq N_{min} & \forall i = 1 \ldots F\ & \forall j = \ldots N \\
     & x_i \geq 0 & \forall i = 1 \ldots F
\end{array}
$$

In Pyomo, this is called an _abstract_ model and it's interesting because the parameters ($c_i, a_{i,j}$) can be added from external files (i.e. the same model can be used to solve a variety of problems). For example, a _concrete_ model will be the following:


$$
\begin{array}{lll}
\max & f(x) =  x_1 + x_2 & \\ 
s.t. & -3x_1 + x_2 \leq 2  & \\
     & x_1 - x_2 \leq 3  & \\
     & x_2 \leq 11  & \\
     & x_1 \leq 6  & \\
     & x_1,x_2 \geq 0
\end{array}
$$

<img src="img/simplex.png" alt="CAChemE" style="width: 300px;"/> 

This problem can be solved using the ** [SIMPLEX Algorithm](https://en.wikipedia.org/wiki/Simplex_algorithm) ** using in the GNU GLPK Solver(enlace).
